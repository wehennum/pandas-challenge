```python
import pandas as pd
from pathlib import Path
```

```python
school_data = Path("../Resources/schools_complete.csv")
student_data = Path("../Resources/students_complete.csv")
```

```python
school_data_df = pd.read_csv(school_data)
school_data_df.head()
```

```python
student_data_df = pd.read_csv(student_data)
student_data_df.head()
```

```python
import pandas as pd

# Calculate the Total Number of Unique Schools
total_unique_schools = school_data_df['School ID'].nunique()

# Calculate the Total Number of Students
total_students = student_data_df['Student ID'].nunique()

# Calculate the Total Budget
total_budget = school_data_df['budget'].sum()

# Calculate the Average Math Score
average_math_score = student_data_df['math_score'].mean()

# Calculate the Average Reading Score
average_reading_score = student_data_df['reading_score'].mean()

# Calculate the Percentage Passing Math
passing_math_percentage = (student_data_df['math_score'] >= 70).mean() * 100

# Calculate the Percentage Passing Reading
passing_reading_percentage = (student_data_df['reading_score'] >= 70).mean() * 100

# Calculate the Percentage Overall Passing
overall_passing_percentage = ((student_data_df['math_score'] >= 70) & (student_data_df['reading_score'] >= 70)).mean() * 100

# Create a DataFrame with the calculated metrics
district_summary_df = pd.DataFrame({
    'Total Schools': [total_unique_schools],
    'Total Students': [total_students],
    'Total Budget': [total_budget],
    'Average Math Score': [average_math_score],
    'Average Reading Score': [average_reading_score],
    '% Passing Math': [passing_math_percentage],
    '% Passing Reading': [passing_reading_percentage],
    '% Overall Passing': [overall_passing_percentage]
})

# Display the district summary DataFrame
district_summary_df
```

```python
import pandas as pd

# Group the student data by school
school_grouped = student_data_df.groupby('school_name')

# Calculate the total students, average math score, and average reading score for each school
school_summary_df = school_grouped.agg({
    'Student ID': 'count',
    'math_score': 'mean',
    'reading_score': 'mean'
})

# Merge the school summary with school data to get the school type and budget
school_summary_df = school_summary_df.merge(school_data_df, on='school_name')

# Calculate the per student budget
school_summary_df['Per Student Budget'] = school_summary_df['budget'] / school_summary_df['Student ID']

# Calculate the percentage passing math for each school
school_summary_df['% Passing Math'] = (school_grouped['math_score'].apply(lambda x: (x >= 70).mean()) * 100).values

# Calculate the percentage passing reading for each school
school_summary_df['% Passing Reading'] = (school_grouped['reading_score'].apply(lambda x: (x >= 70).mean()) * 100).values

# Calculate the percentage overall passing (math AND reading) for each school
school_summary_df['% Overall Passing'] = (school_grouped.apply(lambda x: ((x['math_score'] >= 70) & (x['reading_score'] >= 70)).mean()) * 100).values

# Rename columns for clarity
school_summary_df.rename(columns={
    'school_name': 'School Name',
    'type': 'School Type',
    'Student ID': 'Total Students',
    'budget': 'Total School Budget',
    'math_score': 'Average Math Score',
    'reading_score': 'Average Reading Score'
}, inplace=True)

# Reorder columns
school_summary_df = school_summary_df[['School Name', 'School Type', 'Total Students', 'Total School Budget', 'Per Student Budget', 'Average Math Score', 'Average Reading Score', '% Passing Math', '% Passing Reading', '% Overall Passing']]

# Display the school summary DataFrame
school_summary_df
```
